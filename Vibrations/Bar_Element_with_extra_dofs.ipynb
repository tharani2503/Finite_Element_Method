```python
## import Modules
import numpy as np
import math
from scipy.linalg import eigh
def bar_elem_2nd_ord(n_elems):

    ## constarined DOF
    constrained_dof=[0]

    ## construct local the consistent matrices
    m= np.array([[4,2,-1],
                 [2,16,2],
                 [-1,2,4]])/(30*n_elems)
    k=np.array([[7,-8,1],
                [-8,16,-8],
                [1,-8,7]])/float(3/n_elems)
    
## initialize global mass and stiffness matrices
    M=np.zeros((2*n_elems +1,2*n_elems+1))
    K=np.zeros((2*n_elems +1,2*n_elems+1))

    for i in range (0,2*n_elems-1):
        M_temp=np.zeros((2*n_elems +1,2*n_elems+1))
        K_temp=np.zeros((2*n_elems +1,2*n_elems+1))

        M_temp[i:i+3,i:i+3]=m
        K_temp[i:i+3,i:i+3]=k

        M += M_temp
        K += K_temp

    #print(M,"\n")
    for dof in constrained_dof:
        for i in [0,1]:
            M=np.delete(M,dof,axis=i)
            K=np.delete(K,dof,axis=i)

    eig_vals,eig_vecs=eigh(K,M)
    natural_frequency=np.sqrt(eig_vals)

    return natural_frequency,eig_vecs

## exact frequency
exact_frequency =round(math.pi/2,3)
print(f"exact _frequency == {exact_frequency}")
results =[]

def main()-> None:
    for i in range(1,5):
        frequencies,eig_vecs=bar_elem_2nd_ord(i)
        error=round((exact_frequency - round(frequencies[0],3))/exact_frequency*100,3)
        print("no. of.element::{} >> Fundamental frequency::{} >> Error::{}%\n".format(i,round(frequencies[0],3),error))
        results.append((i,error))


if __name__ == "__main__":

    main()   
```
