```python
import numpy as np
import math
from scipy.linalg import eigh 

def bar_element(n_elems):
    restrained_dofs=[0]
    ## construct local mass and stiffness matrix
    m=np.array([[2,1],
                [1,2]]) / (6*n_elems)
    k=np.array([[1,-1],
                [-1,1]]) * float(n_elems)
    
    ## Initialize matrices
    M=np.zeros((n_elems+1,n_elems+1))
    K=np.zeros((n_elems+1,n_elems+1))
    ## Construct global mass and stiffness matrix
    for i in range(0,n_elems):
        M_temp=np.zeros((n_elems+1,n_elems+1))
        K_temp=np.zeros((n_elems+1,n_elems+1))
        M_temp[i:i+2,i:i+2]=m
        K_temp[i:i+2,i:i+2]=k
        M += M_temp
        K += K_temp

    ## Remove unwanted dofs
    for dof in restrained_dofs:
        for i in [0,1]:
            M=np.delete(M,dof,axis=i)
            K=np.delete(K,dof,axis=i)
    
    ## Solve for EigenValue
    eig_vals,eig_vecs=eigh(K,M)
    frequencies=np.sqrt(eig_vals)
    return M,K,frequencies,eig_vecs

## Analytical solution
exact_frequency=round(math.pi/2,3)
print("exact_frequency::{}".format(exact_frequency))
results=[]

for i in range(1,11):
    M,K,frequencies,eig_vecs=bar_element(i)
    error=round((round(frequencies[0],3) - exact_frequency)/exact_frequency*100,3)
    print("no. of.element::{} >> Fundamental frequency::{} >> Error::{}%\n".format(i,round(frequencies[0],3),error))
    results.append((i,error))
```
